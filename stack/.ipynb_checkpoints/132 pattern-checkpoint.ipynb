We can use various approaches to solve this problem, from brute force O(n^3) solution to O(n) using stacks.

```python
# O (N^2) solution
def find132pattern(nums): 

   #Approach 1: O(N^2)
    minimum = float("inf")
    for i in range(len(nums)):
        minimum = min(minimum, nums[i])
        if minimum == nums[i]:
            continue
        for j in range(len(nums)-1, i, -1):
            if  minimum < nums[j] and nums[j] < nums[i]:
                return True
    return False
print(find132pattern([1,2,3,4]))
```

```python
# O(n) solution using Stack 
def find132pattern(nums): 
    leftMin = []
    leftMin.append(nums[0])
    for i in range(1, len(nums)):
        leftMin.append(min(leftMin[-1], nums[i]))

    stack = []

    for i in range(len(nums)-1, 0, -1):

        if nums[i] <= leftMin[i]:
            continue

        while len(stack) > 0 and stack[-1] <= leftMin[i]:
            stack.pop()

        if len(stack) > 0 and stack[-1] < nums[i]:
            return True

        stack.append(nums[i])

    return False
                
print(find132pattern([1,3,2,4]))
```
