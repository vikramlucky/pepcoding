```python
def stockSpan(arr):
    stack = []
    stack.append((len(arr)-1, arr[-1]))
    
    for i in range(len(arr)-2, -1, -1):
        if  stack[-1][1] > arr[i]:
            stack.append((i, arr[i]))
            continue
        
        while len(stack) > 0 and arr[i] > stack[-1][1]:
            idx, ele = stack.pop()
            arr[idx] = idx - i
        
        stack.append((i, arr[i]))
        
    for i in range(len(stack)):
        p, ele = stack.pop()
        arr[p] = p + 1
    
    return arr
```

```python
arr = [2, 5, 9, 3, 1, 11, 6, 8, 7]
print(stockSpan(arr))
```
