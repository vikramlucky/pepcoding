```python
arr = [5, 5, 3, 8, -2, 7]

def nextGreaterRight(arr):
    stack = []
    stack.append((0, arr[0]))
    
    for i in range(1, len(arr)):
        if  stack[-1][1] > arr[i]:
            stack.append((i, arr[i]))
            continue
        
        while len(stack) > 0 and stack[-1][1] < arr[i]:
            idx, ele = stack.pop()
            arr[idx] = arr[i]
        
        stack.append((i, arr[i]))
        
    for i in range(len(stack)):
        i, ele = stack.pop()
        arr[i] = -1
    
    return arr
print(nextGreaterRight(arr))
        
```

```python

arr = [5, 5, 3, 8, -2, 7]

def nextGreaterLeft(arr):
    stack = []
    stack.append((len(arr)-1, arr[-1]))
    
    for i in range(len(arr)-2, -1, -1):
        if  stack[-1][1] > arr[i]:
            stack.append((i, arr[i]))
            continue
        
        while len(stack) > 0 and stack[-1][1] < arr[i]:
            idx, ele = stack.pop()
            arr[idx] = arr[i]
        
        stack.append((i, arr[i]))
        
    for i in range(len(stack)):
        i, ele = stack.pop()
        arr[i] = -1
    
    return arr
print(nextGreaterLeft(arr))
```

```python

arr = [2 5 9 3 1 12 6 8 7]

def nextGreaterLeft(arr):
    stack = []
    stack.append((len(arr)-1, arr[-1]))
    
    for i in range(len(arr)-2, -1, -1):
        if  stack[-1][1] > arr[i]:
            stack.append((i, arr[i]))
            continue
        
        while len(stack) > 0 and stack[-1][1] < arr[i]:
            idx, ele = stack.pop()
            arr[idx] = idx - i
        
        stack.append((i, arr[i]))
        
    for i in range(len(stack)):
        i, ele = stack.pop()
        arr[i] = i
    
    return arr
print(nextGreaterLeft(arr))
```
