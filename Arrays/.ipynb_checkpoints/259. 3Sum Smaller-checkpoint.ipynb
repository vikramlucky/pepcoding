```python

def threeSumSmaller(nums, target):

    if not nums or len(nums) < 3:
        return 0

    nums.sort()
    answer = 0
    for leftNum in range(len(nums)-2):

        midNum = leftNum + 1
        rightNum = len(nums) - 1

        while midNum < rightNum:

            if nums[leftNum] + nums[midNum] + nums[rightNum] < target:
                answer += rightNum - midNum
                midNum += 1
            else:
                rightNum -= 1
    return answer

```
