```python
# BFS approch: Create a new matrix, or change given matrix set all 1's to large number, 
# append indices of 0s in a queue. Then using BFS approach explore all directions for 0s indices, as long as they
# within bounds, and cell next to has larger values, update that value and append new indices in queue.

from collections import deque
def updateMatrix(mat):
    n = len(mat)
    m = len(mat[0])

    matrix = [[float("inf") for x in range(m)] for k in range(n)]
    queue = deque()

    for i in range(n):
        for j in range(m):          
            if mat[i][j] == 0:
                matrix[i][j] = 0
                queue.append((i,j))
    dirs = [(-1,0), (1,0), (0, -1), (0,1)]

    while queue:      
        i, j = queue.popleft()
        for r, c in dirs:
            new_r = r + i
            new_c = j + c

            if new_r >= 0 and new_r < n and new_c >= 0 and new_c < m and matrix[new_r][new_c] > matrix[i][j] + 1:
                queue.append((new_r, new_c))
                matrix[new_r][new_c] = matrix[i][j] + 1
    return matrix
        
```

```python
mat = [[0,0,0], [0,1,0], [1,1,1]]
print(updateMatrix(mat))
```
